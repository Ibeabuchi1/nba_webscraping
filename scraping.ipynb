```python
from bs4 import BeautifulSoup
import requests
import lxml
```

```python
html_text = requests.get('https://www.jobberman.com/jobs/sales').text
print(html_text)
```

```python
soup = BeautifulSoup(html_text, 'lxml')
```

```python
jobs = soup.find_all('div', class_ = 'mx-5 md:mx-0 flex flex-wrap col-span-1 mb-5 bg-white rounded-lg border border-gray-300 hover:border-gray-400 focus-within:ring-2 focus-within:ring-offset-2 focus-within:ring-gray-500')
```

```python
import pandas as pd
```

```python
result = []

for job in jobs:
    
    posted = job.find('p', class_ = 'ml-auto text-sm font-normal text-gray-700 text-loading-animate').text

    job_title = job.find('p', class_ = 'text-lg font-medium break-words text-link-500').text.strip('\n')
    company_name = job.find('a', class_ = 'text-loading-animate text-loading-animate-link').text.strip('\n')
    location = job.find('span', class_ = 'mb-3 px-3 py-1 rounded bg-brand-secondary-100 mr-2 text-loading-hide').text.strip('\n')
    job_type = job.find_all('span', class_ = 'mb-3 px-3 py-1 rounded bg-brand-secondary-100 mr-2 text-loading-hide')[1].text.strip('\n')
    salary = job.find_all('span', class_ = 'mb-3 px-3 py-1 rounded bg-brand-secondary-100 mr-2 text-loading-hide')[2].text.strip('\n').split()[-1]
    more_info = job.find('div', class_ = 'flex items-center').a['href']
    
    print(f'''
    Company Name: {company_name}
    Job Title: {job_title}
    More info: {more_info}
    
    ''')
    
    print()
```

```python
result = []

for job in jobs:
    posted = job.find('p', class_ = 'ml-auto text-sm font-normal text-gray-700 text-loading-animate').text
    if 'day' in posted:
        job_title = job.find('p', class_ = 'text-lg font-medium break-words text-link-500').text.strip('\n')
        company_name = job.find('a', class_ = 'text-loading-animate text-loading-animate-link').text.strip('\n')
        location = job.find('span', class_ = 'mb-3 px-3 py-1 rounded bg-brand-secondary-100 mr-2 text-loading-hide').text.strip('\n')
        job_type = job.find_all('span', class_ = 'mb-3 px-3 py-1 rounded bg-brand-secondary-100 mr-2 text-loading-hide')[1].text.strip('\n')
        salary = job.find_all('span', class_ = 'mb-3 px-3 py-1 rounded bg-brand-secondary-100 mr-2 text-loading-hide')[2].text.strip('\n').split()[-1]

        result.append((job_title, company_name, location, job_type, salary, posted))
    
#         print(f'''
#         Company Name: {company_name}
#         Job Title: {job_title}
#         Job Type: {job_type}
#         Location: {location}
#         Salary: {salary}
#         Posted: {posted}
#         ''')
    
#     print('############')
```

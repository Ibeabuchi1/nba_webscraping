```python
from bs4 import BeautifulSoup
import requests
import lxml
import pandas as pd
```

```python
url_start = "https://www.basketball-reference.com/awards/awards_{}.html"
```

```python
years = list(range(1991, 2022))
```

```python
# download html pages to local
for year in years:
    url =  url_start.format(year)
    data = requests.get(url)
    
    with open("mvp/{}.html".format(year), "w", encoding="utf-8") as f_in:
        f_in.write(data.text)        
```

```python
#  read html pages
with open("mvp/2012.html", "r", encoding="utf-8") as f_out:
    page = f_out.read()
```

```python
df = []

for year in years:
    with open("mvp/{}.html".format(year), "r", encoding="utf-8") as f_out:
        page = f_out.read()
    
    soup = BeautifulSoup(page, "html.parser")
    soup.find('tr', class_='over_header').decompose()
    mvp_table = soup.find(id='mvp')
    mvp = pd.read_html(str(mvp_table))[0]
    mvp['year'] = year
    
    df.append(mvp)
```

```python
mvp = pd.concat(df)
```

```python
mvp.head()
```

```python
# save dataframe to csv format
mvp.to_csv("mvps.csv")
```
